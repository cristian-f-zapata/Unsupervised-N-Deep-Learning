# Mean Shift Clustering

```python
# Surpress any warnings:
def warn(*args, **kwargs):
    pass
import warnings
warnings.warn = warn

import numpy as np
import cv2 as cv
from sklearn.cluster import MeanShift, estimate_bandwidth
from mpl_toolkits import mplot3d
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import math
```

```python
img = cv.imread('frutas.jpg')
plt.imshow(cv.cvtColor(img, cv.COLOR_BGR2RGB))
plt.show() 
```

```python
img = cv.medianBlur(img, 7)
plt.imshow(cv.cvtColor(img, cv.COLOR_BGR2RGB))
```

```python
ax = plt.axes(projection ="3d")
ax.scatter3D(img[:,:,0],img[:,:,1],img[:,:,2])
ax.set_title('Pixel Values ')
plt.show()
```

```python
img.shape
```

```python
X = img.reshape((-1,3))
print("shape: ",X.shape)
print("data type   : ",X.dtype)
```

```python
X = np.float32(X)
```

## Aplicamos Mean Shift a la imagen

```python
bandwidth = estimate_bandwidth(X, quantile=.06, n_samples=3000)
bandwidth 
```

```python
ms = MeanShift(bandwidth=bandwidth,bin_seeding=True)
ms.fit(X)
```

```python
labeled=ms.labels_
labeled
```

```python
clusters=ms.predict(X)
clusters
```

```python
np.unique(labeled)
```

```python
ms.cluster_centers_
```

```python
cluster_int8=np.uint8(ms.cluster_centers_)
cluster_int8
```

```python
ms.predict(X)
```

```python
ax = plt.axes(projection ="3d")
ax.scatter3D(img[:,:,0],img[:,:,1],img[:,:,2])
ax.set_title('Pixel Values ')
plt.show()

ax = plt.axes(projection ="3d")
ax.set_title('Pixel Cluster Values  ')
ax.scatter3D(cluster_int8[:,0],cluster_int8[:,1],cluster_int8[:,2],color='red')
plt.show()
```

```python
result=np.zeros(X.shape,dtype=np.uint8)

for label in np.unique(labeled):
    result[labeled==label,:]=cluster_int8[label,:]    
    

result=result.reshape(img.shape)
```

```python
plt.imshow(cv.cvtColor(result, cv.COLOR_BGR2RGB))
plt.show()
```

```python
for label in np.unique(labeled):
    result=np.zeros(X.shape,dtype=np.uint8)
    result[labeled==label,:]=cluster_int8[label,:]  
    plt.imshow(cv.cvtColor(result.reshape(img.shape), cv.COLOR_BGR2RGB))
    plt.show()
```
