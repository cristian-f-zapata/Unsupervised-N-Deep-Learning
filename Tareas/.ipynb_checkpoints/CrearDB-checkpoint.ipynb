# <span style='color:purple'>**TRANSFORMACIONES A LAS IMAGENES** $=$ **CREACIÓN DE BASE DE DATOS**<span/>

# 1. Convertimos todas las fotos a un mismo tamaño, y convertimos a escala de grises

```python
!pip install keras==2.3.1
```

```python
import numpy as np
#import tensorflow as tf



from keras_preprocessing.image import load_img
#from tensorflow.keras.utils import load_img
from keras_preprocessing.image import img_to_array
from PIL import Image
from PIL import Image, ImageOps, ImageFilter
```

```python
imgs=list()

labels=list()

#se reescalan las fotos tal que queden de 128X128
dim=128

for i in range(n_muestras):
    #Estamos cargando las imagenes con keras, le estamos cambiando el tamaño
    #las estamos poniendo en escalas de grises
    #a la carpeta imagenes estamos añadiendo esas imagenes pero como arrays
    #Y les estamos poniendo unas etiquetas, donde los 1 serán para los perros
    #y los ceros para los gatos
    img=load_img((rutas_perros[i]), target_size=(dim,dim), grayscale=True)
    imgs.append(img_to_array(img))
    labels.append(1)
    
    img=load_img((rutas_gatos[i]), target_size=(dim,dim),grayscale=True)
    imgs.append(img_to_array(img))
    labels.append(0)

array_imgs=np.asarray(imgs)
array_labels=np.asarray(labels)
```

```python
array_labels[0]#la primera imagen me la cataloga de perro.

array_imgs[0] #Como vemos, la primera imagen está convertida en array (de 128 filas, y 128 columnas)
```

Aquí la imagen está convertida a escala de grises

```python
plt.imshow(array_imgs[255])
```

Se normaliza para que las intensidades de todas las imagenes tengan un máximo de 255 y mínimo de 0

```python
def normalize_scale(image_data):
    scale_min=0
    scale_max=255
    return((image_data-scale_min)/(scale_max-scale_min))
```

```python
array_imgs_scale=normalize_scale(array_imgs)#dataset contruido
```

```python
array_imgs_scale[0]
```

```python
plt.imshow(array_imgs_scale[255])
```

```python
array_imgs_scale_v2=array_imgs_scale.reshape(n_muestras*2,-1)\
.reshape(n_muestras*2,dim,dim)
```
