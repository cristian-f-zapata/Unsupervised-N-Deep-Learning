# Ajuste de la Base de Datos

A fin de que sea computacionalmente menos exigente, se normaliza la imagen (los colores). 
Y adem√°s, se ajusta el tama√±o. Tal que todas las imagenes tengan el mismo tama√±o.

```python
import numpy as np
#import tensorflow as tf



from keras_preprocessing.image import load_img
#from tensorflow.keras.utils import load_img
from keras_preprocessing.image import img_to_array
from PIL import Image
from PIL import Image, ImageOps, ImageFilter
```

```python
imgs=list()

labels=list()

#se reescalan las fotos tal que queden de 128x128
dim=128

for i in range(306):
    img=load_img((rutas_women[i]), target_size=(dim,dim), grayscale=True)
    imgs.append(img_to_array(img))
    labels.append(0)
    
    img=load_img((rutas_men[i]), target_size=(dim,dim), grayscale=True)
    imgs.append(img_to_array(img))
    labels.append(1)
    
array_imgs=np.asarray(imgs)
array_labels=np.asarray(labels)
```
